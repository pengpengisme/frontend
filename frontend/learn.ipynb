React 運作流程
把要渲染的內容丟入ReactDOM -> ReactDOM整理要渲染的內容 -> 對原始綁定的div在DOM上的位置做渲染 -> 完成畫面渲染

React 採用JSX語法(特殊的Javascript語法)
    1. html語法可以當作參數傳遞
    2. 只能傳遞一個元素:有一堆div、button......要傳的話，要用一個container把他們包起來。
    3. 可以在 html 標籤中利用「{}」寫javascript 表示式
    4. style變為一物件、屬性名稱規則改用駝峰法(用大寫區隔)、屬性的值變成字串
        const styleArgument = { fontSize: '100px', color: 'red' };

        ReactDOM.render(
            <h1 style = { styleArgument } > Hello, world! </h1>,
            document.getElementById('root')
        );

        直接在html標籤中給style值的寫法:要用「{ { } }」，外面那層括號代表裡面是javascript語法，裡面的括號表示物件型態。
        ReactDOM.render(
            <h1 style = {{ fontSize: '100px', color: 'red' }} > Hello, world! </h1>,
            document.getElementById('root')
        );
    5. 「class」屬性變成「className」
        ReactDOM.render(
            <h1 className = "title" > Hello, world! </h1>,
            document.getElementById('root')
        );
    6. 在元素上傳遞屬性時若省略要指定的值，該資料會獲得true布林值
        ex:<button value > 是 </button> = <button value={true} > 是 </button>
    7. 函式綁定
        [第一種]
        <button value={true}  onClick={ myFunction }>是</button>

        函式只會接收到一個event類別的參數，並不能傳遞其他參數，如果我們要取得元素的value，就必須要透過原生DOM的event.target.value來取得
        const getValue=(event)=>{
            console.log(event.target.value)
        }

        ReactDOM.render(
            <div>
                <button value={true} onClick={getValue}>按下以取得數值</button>
            </div>,
            document.getElementById('root')
        );

        [第二種] : 想傳什麼參數、使用幾個函式、做什麼運算都可以
        <button value="true"  onClick={(e)=>{ myFunction1();myFunction2()} }>是</button>

        因為是在render創造一個新函式，每次渲染時都會創造一次，會影響效能，所以一般能用1的話就會用1的方法

        const getValue=(value)=>{
            console.log(value)
        }

        ReactDOM.render(
            <div>
                <button value={true} onClick={(event)=>{getValue(event.target.value)}}>
                    按下以取得數值 
                </button>
            </div>,
            document.getElementById('root')
        );

React component語法:元素名稱第一個字母必須要是大寫、和函式(或class)名稱相同，JSX才會知道這是自製的component
這個語法讓我們可以把程式碼以函式(或是class，後面會提到)包成模組，變成一個自己訂作的元素，然後用跟使用button、div這些元素一樣的語法在JSX中使用，程式碼中的{函式名稱}變成了<函式名稱/>。

    

什麼是props?
component當中所有的attribute(ex:id, value, name)和其他用來控制元件的參數(像是button中的文字)整合成一個物件，稱為props

使用props綁定資料的資料型態問題
<App number="87" getData="true"/>
-->> App接到的number和getData都是字串，並不是整數和布林值

必須要加上{}才會接到正確的js資料型態，使用變數時也要這樣
<App number={87} getData={true}/>

當使用同class scope的其他member function時，必須要使用this.函式名稱 = this.函式名稱.bind(this)綁定此物件。

import React, { Component } from 'react';

class App extends Component{
    constructor(props) { 
        super(props);
        this.name="舊的名字"; // 加入name
        this.changeName=this.changeName.bind(this); //綁定至自己
    }

    changeName(newName){ // 定義changeName，同在同一個class中
        this.name=newName;
        console.log("hey")
    }
}
